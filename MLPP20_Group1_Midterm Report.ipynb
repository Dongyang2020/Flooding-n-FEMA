#Background

Researchers investigate contributions of *flood risk* in an effort to decrease the future impacts of flooding. In the natural hazards research community, [risk at-large is often characterized](https://www.ipcc.ch/site/assets/uploads/2018/03/SREX-Chap2_FINAL-1.pdf) as a combination of the type of hazard in question, the amount of exposure in a community, and the overall vulnerability. This can also be thought of as the [conceptual equation](https://www.preventionweb.net/risk/vulnerability) shown below. **(GGD: While I like the concept, I would instead phrase this as "Risk depends on three factors: hazard, exposure, and vulnerability".  ***Risk = hazard * exposure  * vulnerability**
 
In this equation, a hazard is usually classified as a weather event (storm, hail, snow, etc.), climatic event (such as drought), or geophysical phenomenon (earthquakes, landslides, etc.). Hazards can also include [wars, pandemics, and industrial accidents](https://www.ifrc.org/en/what-we-do/disaster-management/about-disasters/definition-of-hazard/). Exposure refers to the amount of structures, infrastructure, and capital within a community that is threatened when a hazard event occurs. For example a densely-populated city will have larger exposure as compared to a rural town. The overall vulnerability of a community refers to its [capacity to handle and respond to hazard events](https://www.ifrc.org/en/what-we-do/disaster-management/about-disasters/what-is-a-disaster/what-is-vulnerability/) and return as a functioning society. For example, a community with high vulnerability could have a fragile building stock, weak infrastructure systems, poor emergency response systems, few financial resources, and/or large economic disparities in its population. A community with low vulnerability, or a robust community, is able to quickly respond to hazard events and all community members, regardless of socio-economic status, are minimally impacted. 

##Flood Risk Characterization and Social Impacts

While the hazard component of flood risk (either a flood or storm event) cannot be altered since it is a geophysical phenomenon, the level of exposure and vulnerability related to flooding hazards can be adjusted to minimize the overall flood risk. The relatively high level of exposure of US residents to flood hazards stems from the early development of towns [near rivers](https://www.nationalgeographic.org/encyclopedia/flood-plain/) as the close proximity allowed for the transportion of goods by boat and provided drinking water and irrigation. However, as populations grew and infrastructure was added, the exposure to flood risk often increased. The [NYU Furman Center](https://furmancenter.org/files/Floodplain_PopulationBrief_12DEC2017.pdf) has reported that nearly ten percent of the U.S. population on-average lived in floodplains between the years 2011 and 2015. In particular, there are three states with a significantly large portion of their population living within floodplains, with Arizona at 64%, Florida at 26%, and North Dakota at 20%. Additionally, as regions become more urbanized, and more surfaces are converted from pervious surfaces to paved surfaces, the area becomes at-risk of flooding, specifically flash flood events. This is because the water in an urban environment, that would normally drain into the ground, must instead travel across the paved surface usually into a piped drainage system. During heavy rainfall, water can quickly accumulate on paved surfaces and oftentimes, drainage systems cannot handle the increased volume of water. This can lead to urban flood damages. Between 1993 and 2017, Wisconsin and Iowa experienced the most number of urban flooding instances, with [3,663 urban flooding events in total](https://cdr.umd.edu/sites/cdr.umd.edu/files/resource_documents/COMPRESSEDurban-flooding-report-online-compressed-0319.pdf).

##FEMA Programs
FEMA is the government agency largely leading the national effort in reducing national flood risk. The agency is housed within the Department of Homeland Security and has an [annual operating budget of $28.7 billion USD](https://en.wikipedia.org/wiki/Federal_Emergency_Management_Agency). The agency manages a variety of programs to provide assistance after disaster events occur and implements pre-disaster mitigation programs to reduce overall risk. This report focuses on three key programs within FEMA that are meant to reduce the nation’s flood risk: the National Flood Insurance program (NFIP), the Community Rating System (CRS), and the Hazard Mitigation Grants Program (HMGP). 

###National Flood Insurance Program

One method for decreasing vulnerabilities to future flood events is through flood insurance. Flood insurance serves to quantify risk for the consumer and allows individuals to receive financial assistance following a damaging flood event. Although Congress brought flood control into the federal purview in 1917, it wasn't until the 1950s that **(GGD: President)** Truman introduced a national flood re-insurance program. The private flood insurance industry had crumbled under the weight of the 1927 Great Mississippi River Flood, and until only recently, continually maintained that it was "not commercially feasible". With the creation of the National Flood Insurance Program(NFIP) in 1968, qualified companies began to sell subsidized policies to property owners in flood-prone areas. Since then, the program has evolved through the use of flood-maps and community investments, but through a series of profoundly damaging storms and historically low policy rates, the accounting in 2019 showed nearly $21 billion of debt. **(GGD: Reference?)** Today, the NFIP is housed under FEMA, and while NFIP has some resemblence to the commercial enterprise of risk management, as one researcher put it, it is more often "a means to have floodplain occupants partially pre-pay the costs of their disaster aid" **(GGD: Reference?)**.

Understanding that insurance is a primary strategy by which the U.S. has addressed climatic risks, this report will explore the equitability **(GGD: defined in what way?)** of NFIP's resources. While over 98% of the U.S. population are eligible to purchase NFIP policies, only 5 million households had flood insurance in 2018. Additionally, 2018 NFIP Affordability Framework revealed that "policyholders have higher median incomes than non-policyholders...for 50 out of 52 states tested". One study on the determinants of enrollment found "income, education, age, percentage of white population, and percentage of owner occupied homes" to positively correlate with take-up rates. **(GGD: Reference?)** In particular, a University of Maryland & Texas A&M study revealed that access to flood insurance policies is divided along socio-economic status. ###Community Rating System

Linked to the NFIP is the [Community Rating System](https://www.fema.gov/national-flood-insurance-program-community-rating-system) (CRS), which is a credit-based incentive program that promotes active flood risk management in local communities. Participating communities are scored based on a [list of activities](https://www.fema.gov/media-library-data/1562876385763-262f88e35f4dde1484be9093b099e670/2018_NFIP_CRS_Brochure_June_2018_508OK_corrected.pdf) such as flood insurance promotion, open space preservation, stormwater management, and other activities. Participating communities can have discounted flood insurance rates of up to 45 percent for policies located within the 100-year floodplain, depending on the number of activities implemented by the community. Critiques of the CRS vary, with [one study in 2017](https://onlinelibrary-wiley-com.udel.idm.oclc.org/doi/full/10.1111/risa.12853) specifically investigating the question: what is the effect of the CRS on poverty and income inequality? The study found that median incomes are lower in CRS communities at-large, but rise specifically in floodplains. **(GGD: just to be clear, there are CRS communities that are not in floodplains? Overall, while the CRS program encourages communities to decrease their flood risk holistically through a range of activities, the outcomes of the program are not always equitable.

###Hazard Mitigation Grant Program

The core FEMA program that supports flood risk reduction activities is the [Hazard Mitigation Grant Program](https://www.fema.gov/hazard-mitigation-grant-program) (HMGP). This program supports a range of activities including structural retrofitting, flood diversion and green infrastructure projects, and acquisition of risky properties (also known as ‘buyouts’). These grants require federal, state, and local coordination, where the federal government will pay for 75 percent of these projects and state or local governments will pay for the remaining 25 percent of the cost. A [recent analysis](http://d279m997dpfwgl.cloudfront.net/wp/2018/07/Kousky-and-Shabman-2017.pdf) showed that "almost 90% of flood risk reduction funding comes after a big flood". While post-disaster needs are immense, the same study pointed to the downsides of such large, reactive measures that often have less accountability than "pre-flood risk reduction programs". The latter presents a challenge to policy makers as they plan and budget to reduce risk. 
 
In particular, ["Repetitive loss properties"](https://www.nrdc.org/media/2017/170724) pose a particular risk to communities and also the NFIP program. These properties have been identified as a disproportionately costly category of claims, with 0.6% of all NFIP insured homes accounting for 9.6% of all damages paid. This issue lies at the convergence of many quantitative and qualitative circumstances, with one study identifying "finance, risk perception, place attachment, and trust" as factors that homeowners must wrangle with in deciding to stay in their home, or move to a different location.  These characteristics highlight the challenges of the buyout program within the HMGP and perhaps explains why only [43,000 properties in the past 30 years](https://www.nrdc.org/sites/default/files/going-under-post-flood-buyouts-report.pdf) have participated in the buyout program. **(GGD: and how many were elligible?)**
 
While [some studies](https://rucore.libraries.rutgers.edu/rutgers-lib/49079/) have evaluated equity aspects of post-disaster assistance programs, such as FEMA’s individual assistance program, few have reviewed equity aspects of the HMGP program and whether vulnerable communities are receiving appropriate assistance. One recent study specifically reviewed the buyout program within the HMGP and found that [richer counties were receiving more assistance for property buyouts](https://advances.sciencemag.org/content/5/10/eaax8995), as compared to poorer counties. Overall though, it is largely unknown whether FEMA mitigation programs are adequately serving vulnerable and disadvantaged communities. Thus, FEMA’s effectiveness in reducing the nation’s overall flood risk still requires further investigation.

```
# -- accumulate sum of missing damages
missing_dmg = 0

# -- loops through csv list for every year's file in noaa's website 
for i in range(len(csv)):
  # damage year column header as a string
  dmgyr = "dmg"+str(year[i])

  # read the noaa csv file for the given year
  noaa_all = pd.read_csv(csv[i], skiprows=0)

  # pull data from storms events that influence flood damage
  noaa_all = noaa_all[(noaa_all["EVENT_TYPE"] == "Coastal Flood") | 
                    (noaa_all["EVENT_TYPE"] ==  "Flash Flood") | 
                    (noaa_all["EVENT_TYPE"] == "Flood") | 
                    (noaa_all["EVENT_TYPE"] == "Heavy Rain") | 
                    (noaa_all["EVENT_TYPE"] == "High Wind") | 
                    (noaa_all["EVENT_TYPE"] ==  "Hurricane") | 
                    (noaa_all["EVENT_TYPE"] == "Lakeshore Flood") | 
                    (noaa_all["EVENT_TYPE"] == "Marine High Wind") | 
                    (noaa_all["EVENT_TYPE"] ==  "Marine Strong Wind") | 
                    (noaa_all["EVENT_TYPE"] == "Marine Thunderstorm Wind") | 
                    (noaa_all["EVENT_TYPE"] == "Seiche") | 
                    (noaa_all["EVENT_TYPE"] == "Storm Surge/Tide") | 
                    (noaa_all["EVENT_TYPE"] == "Thunderstorm Wind") | 
                    (noaa_all["EVENT_TYPE"] == "Tropical Depression") | 
                    (noaa_all["EVENT_TYPE"] == "Tropical Storm")]

  # extract state FIPS, county name, county FIPS, and property damage values
  noaa_all = noaa_all[['STATE_FIPS','CZ_TYPE','CZ_FIPS','CZ_NAME','DAMAGE_PROPERTY']].copy()

  # -- removes Puerto Rico, Virgin Islands, American Samoa, Guam, Northern Mariana Islands, and all bodies of water (ATLANTIC NORTH, ATLANTIC SOUTH,
  # -- E PACIFIC, GULF OF MEXICO, HAWAII WATERS, LAKE HURON, LAKE ONTARIO, LAKE MICHIGAN, LAKE ERIE, LAKE SUPERIOR, LAKE ST CLAIR, )
  noaa_all= noaa_all[noaa_all["STATE_FIPS"].astype(int) < 57]
  #print(noaa_all['STATE_FIPS'].astype(int).max())

  # convert DAMAGE_TYPE values from string to int
  noaa_all[dmgyr] = (noaa_all['DAMAGE_PROPERTY'].replace(r'[KMB]+$', ' ', regex=True).replace(' ',np.nan).astype(float) * noaa_all['DAMAGE_PROPERTY'].str.extract(r'[\d\.]+([KM]+)',expand=False).fillna(1).replace(['K','M','B'],[10**3,10**6,10**9]).astype(int))
  # -- Drop storms with na in damage column
  noaa_all = noaa_all.dropna(subset=[dmgyr])

  # construct a state+county ID, so all counties have unique identifier (combines state + county/zone code)
  state = noaa_all["STATE_FIPS"].astype(int).astype(str)
  county = noaa_all["CZ_FIPS"].astype(int).astype(str)
  county = "00"+ county 
  county = county.str[-3:]
  geoid = (state + county).astype(int)
  noaa_all["STATE_CO_FIPS"] = geoid

  # -- dataframe with just zone storm events
  noaa_z = noaa_all[noaa_all['CZ_TYPE'] == 'Z'][['STATE_CO_FIPS',dmgyr,'CZ_NAME']]

#-------------------------------------------------------------------------------------
###NOAA zone fips to zone->county county fips

  # group by zone FIPS and sum total damages per county for this year
  noaa_zg = noaa_z.groupby("STATE_CO_FIPS").sum().reset_index()

  # -- co_fips["FIPS"] not in "STATE_CO_FIPS" ~~~~~~these are the noaa zone FIPS that don't match with the zone->county zone fips
  noaa_zg[~noaa_zg["STATE_CO_FIPS"].isin(ztoc['zone_fips'])]["STATE_CO_FIPS"].count()

  # -- merge noaa data with zone-to-county dataframe on state+zone fips
  noaa_zm = noaa_zg.merge(ztoc,left_on="STATE_CO_FIPS",right_on='zone_fips',how='left') 

  #This calculates the amount of damages that get dropped because the zone fips don't match
  noaa_zm_null = noaa_zm[noaa_zm['state'].isnull()]
  noaa_zm_null_sum = noaa_zm_null[dmgyr].sum()
  missing_dmg += noaa_zm_null_sum

  # -- only select county fips and damage columns
  noaa_zm = noaa_zm[['c_fips',dmgyr]]
  # -- rename county fips column
  noaa_zm.columns = ['CO_FIPS',dmgyr]

#-------------------------------------------------------------------------------------

  # -- dataframe with just county storm events
  noaa_c = noaa_all[noaa_all['CZ_TYPE'] == 'C'][['STATE_CO_FIPS',dmgyr]]
  # -- renmae fips column
  noaa_c.columns = ['CO_FIPS',dmgyr]

  # -- stack county and zones dataframe to combine the data back together
  noaa_all2 = pd.concat([noaa_c,noaa_zm])

  # group by county FIPS with the zone->county FIPS and sum total damages per county for this year
  noaa_all2g = noaa_all2.groupby("CO_FIPS").sum().reset_index()
  noaa_all2g

  # merge onto compiled dataframe
  #noall = co_fips.reset_index().merge(noaa_all2g,left_on="FIPS", right_on="CO_FIPS", how ='outer').set_index('index')
  noall_shp = co_fips.merge(noaa_all2g,left_on="FIPS", right_on="CO_FIPS", how ='outer')
  co_shp[dmgyr] = noall_shp[dmgyr].copy() 
```

#### Fit per capita Damage with Total Population

```
# -- get arrays of per capita damage and population data
dmg = census_2018_geo['dmg_sum'].agg(np.log10).values
pop = census_2018_geo['total_pop'].agg(np.log10).values
# so that they show normal distribution

dmgp = dmg/pop
```

```
# -- standardized all variables
pop_st =(pop - pop.mean())/pop.std()
dmgp_st = (dmgp - dmgp.mean())/dmgp.std()
dmg_st = (dmg - dmg.mean())/dmg.std()
```

```
# -- initialize an SGD regression model
lm_pop_dmgp = SGDRegressor(verbose=1)

# -- fit total population with per capita damage
lm_pop_dmgp.fit(pop_st.reshape(-1, 1), dmgp_st)

# -- unstandardize the model
model_pop_dmgp_unst = model_pop_dmgp * dmgp.std() + dmgp.mean()
```

```
# -- plot the unstanderdized fit model
fig, ax = plt.subplots(figsize=[7, 7])
ax.scatter(pop, dmgp)
ax.plot(pop, model_pop_dmgp_unst, color="darkred")
ax.set_xlabel("Population (log)")
ax.set_ylabel("Per Capita Damage by Flood Disaster")
ax.set_title("Correlation between Per Capita Damage and Total Population in US counties")
fig.show()
# fig.savefig("cor_capitadamage&pop_unst.png")
```

#### Fit Total Damage with Total Population

```
# -- initialize an SGD regression model to fit total population with total damage
lm_pop_dmg = SGDRegressor(verbose=1)

lm_pop_dmg.fit(pop_st.reshape(-1, 1), dmg_st)

# -- "predict" the data
model_pop_dmg = lm_pop_dmg.predict(pop_st.reshape(-1, 1))
```

```
# -- unstandardize the model
model_pop_dmg_unst = model_pop_dmg * dmg.std() + dmg.mean()
# -- plot the unstanderdized fit model
fig, ax = plt.subplots(figsize=[7, 7])
ax.scatter(pop, dmg)
ax.plot(pop, model_pop_dmg_unst, color="darkred")
ax.set_xlabel("Population (log)")
ax.set_ylabel("Total Damage by Flood Disaster (log(US Dollar))")
ax.set_title("Correlation between Total Damage and Total Population in US counties")
fig.show()
# fig.savefig("cor_capitadamage&pop_unst.png")
```

```
# calculate MSE of this model
mse_pop = ((model_pop_dmg_unst-dmg)**2).sum()
print('Mean Squared Error of this model:',"{:.2f}".format(mse_pop))
```

The data for the NFIP is sourced from the [Federal Insurance & Mitigation Administration National Flood Insurance Program (FIMA NFIP) Redacted Claims Dataset](https://www.fema.gov/media-library/assets/documents/180374), which includes data about flood insurance policies, the damage claims by policy-holders and also the CRS. Data since 1996 has been included in this analysis. The insurance claims sums the total claims for building damages, contents damages, and increased cost of compliance. The total insurance coverage data, available in the last column in the table, combines total building and contents coverage. CRS data is provided in a later section. The data dictionary for the new variables presented in this section are also available below.
 
The figures show a difference in insurance claim payments and insurance coverages across the US, using a log scale. A log scale was used to better highlight the counties with significant insurance coverage and claims. From the figure it can be seen that although data is not completed in the Central US, the Southeast has more money paid by FEMA.
